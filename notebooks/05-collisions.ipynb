```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.rcParams["font.family"] = "Linux Libertine O"
plt.rcParams["font.size"] = 30
```

```python
## Plotting the unslotted Aloha probability
L = 5.0
T = 30000.0
N = np.arange(0.0,220.0,10.0,)
G = 4 * N * L / T
P = np.exp(-2*G)
```

```python
f, ax = plt.subplots(figsize=(18, 5))

plt.plot(N, P, linewidth=5.0)
plt.xlabel('Number of cars')
plt.ylabel('P(Success)')

plt.xlim([0,210])
plt.ylim([0.7, 1.0])

plt.grid(True)

plt.savefig('../pics/collisions.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
